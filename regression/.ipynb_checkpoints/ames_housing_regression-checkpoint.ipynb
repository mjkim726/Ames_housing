```python
def eda(dataframe):
    dataframe = dataframe[(ames_train['neighborhood'] !='GrnHill') & (ames_train['neighborhood'] != 'Landmrk')]
    
    #Dropping everything above 20% null
    dataframe.drop(['pool_qc', 'misc_feature', 'alley', 'fence', 'fireplace_qu'], axis = 1, inplace = True)
    
    #fill in the lot frontage by neighborhood and lot shape
    dataframe['lot_frontage'] = dataframe.groupby(['neighborhood', 'lot_shape'])['lot_frontage'].transform(lambda x: x.fillna(x.median()))
    dataframe['lot_frontage'] = dataframe.groupby(['neighborhood'])['lot_frontage'].transform(lambda x: x.fillna(x.median()))
    
    #garage quality
    dataframe['garage_qual'].fillna('NA', inplace = True)
    
    #encoding exterior condition
    extcon_qual_encode = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1
    }
    dataframe['exter_cond'] = dataframe['exter_cond'].map(extcon_qual_encode)
    
    #encode heating to gas or no gas
    dataframe['heating'] = dataframe['heating'].apply(lambda x: 0 if 'Gas' in x else 1)
    
    #encoding heating quality
    heating_qual_encode = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1
    }
    dataframe['heating_qc'] = dataframe['heating_qc'].map(heating_qual_encode)
    
    #encoding kitchen quality
    kitchen_qual_encode = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1
    }
    dataframe['kitchen_qual'] = dataframe['kitchen_qual'].map(kitchen_qual_encode)
    
    #encoding ac
    dataframe['central_air'] = dataframe['central_air'].apply(lambda x: 0 if x =='Y' else 1)
    
    #encoding electrical into breaker or no breaker
    dataframe['electrical'] = dataframe['electrical'].apply(lambda x: 0 if x =='SBrkr' else 1)
    
    #drop columns
    dataframe.drop([col for col in dataframe.select_dtypes(include = ['object']) if (col != 'ms_zoning') & (col != 'lot_shape') & \
     (col != 'land_contour') & (col != 'neighborhood') & (col != 'bldg_type') & (col != 'house_style') & (col != 'garage_qual')], axis = 1, inplace = True)
    
    #change ms subclass into objects
    dataframe['ms_subclass'] = dataframe['ms_subclass'].apply(lambda x: str(x))
    
    #fill in all the numerical types with 0 because the houses don't have that feature
    dataframe = dataframe.apply(lambda x: x.fillna(0) if x.dtype.kind in 'biufc' else x)
    
    #combine bathrooms
    dataframe['grl_bath'] = dataframe['full_bath'] + 0.5*dataframe['half_bath']
    dataframe['bsmt_bath'] = dataframe['bsmt_full_bath'] + 0.5*dataframe['bsmt_half_bath']
    
    #combine porch area
    dataframe['porch'] = dataframe['wood_deck_sf'] + dataframe['open_porch_sf'] + dataframe['enclosed_porch'] + dataframe['3ssn_porch'] + dataframe['screen_porch']
    
    #drop columns we don't need
    dataframe.drop(['id', 'pid', 'full_bath', 'half_bath', 'bsmt_full_bath', 'bsmt_half_bath', 'open_porch_sf',\
                   'enclosed_porch', '3ssn_porch', 'screen_porch', 'wood_deck_sf', 'misc_val'], axis = 1, inplace = True)
    
    return pd.get_dummies(dataframe, columns = dataframe.select_dtypes(include = ['object']).columns)
```

```python
train = eda(ames_train)
```

### Model

```python
features = [col for col in train.columns if (col!='saleprice')]
X = train[features]
y = train['saleprice']
```

```python
X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size = .3, random_state = 101)
```

```python
ss = StandardScaler()
pca = PCA(.9)
knr = KNeighborsRegressor()
skb = SelectKBest()
lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()
```

```python
pipe = Pipeline([
    ('ss', ss),
    ('pca', pca),
    ('lr', lr)
])
```

```python
params = {

}

gslm = GridSearchCV(pipe, cv = 5, param_grid = params, verbose = 1, n_jobs = -1)
gslm.fit(X_train, y_train)
print('The best parameters: ', gslm.best_params_)
print('The Best score: ', gslm.best_score_)
print('The holdout score:', gslm.score(X_holdout, y_holdout))
```

```python
pipe = Pipeline([
    ('ss', ss),
    ('pca', pca),
    ('lasso', lasso)
])
```

```python
params = {
    'lasso__alpha': np.linspace(530, 600)
}

gsl = GridSearchCV(pipe, cv = 5, param_grid = params, verbose = 1, n_jobs = -1)
gsl.fit(X_train, y_train)
print('The best parameters: ', gsl.best_params_)
print('The Best score: ', gsl.best_score_)
print('The holdout score:', gsl.score(X_holdout, y_holdout))
```

```python
pipe = Pipeline([
    ('ss', ss),
    ('pca', pca),
    ('skb', skb),
    ('sfm', SelectFromModel(Lasso())),
    ('knr', knr)
])
```

```python
params = {'sfm__estimator': [Lasso(alpha=a, random_state=101) for a in np.linspace(500, 600)],
          'knr__n_neighbors': np.arange(5,10),
          'skb__k': np.arange(10, 60, 10)
         }
gsl = GridSearchCV(pipe, cv = 10, param_grid=params, verbose = 1, n_jobs = -1)
gsl.fit(X_train, y_train)
print('The best parameters: ', gsl.best_params_)
print('The Best score: ', gsl.best_score_)
print('The holdout score:', gsl.score(X_holdout, y_holdout))
```
